#Using Beautiful Soup Library

1.   We will find Page Title
2.   All links that are present on the website

```python
import requests
from bs4 import BeautifulSoup

# Make a request to the website
url = 'https://www.thaparsummerschool.com/schedule'

response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

#1
# Find and extract specific elements from the HTML
title = soup.title.text
print('Page title:\n', title, '\n\n\n')

#2
# Find all the links on the page
links = soup.find_all('a')
print('Links:')
for link in links:
    print(link.get('href'))
```

3. Now we will find all the raw data present on the website.


```python
# import requests
# from bs4 import BeautifulSoup

# # Make a request to the website
# url = 'https://www.thaparsummerschool.com/schedule'
# response = requests.get(url)

# # Create a BeautifulSoup object
# soup = BeautifulSoup(response.text, 'html.parser')

# Find all text in the HTML
text_elements = soup.find_all(text=True)

# Store the extracted text in a variable named raw_data
raw_data = ""
for element in text_elements:
    if element.parent.name not in ['script', 'style']:
        raw_data += element.strip() + "\n"

```

```python
# Print the raw_data
print(raw_data)
```
